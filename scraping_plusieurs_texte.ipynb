```python
# Importer les packages
import nltk
import pandas as pd
from newspaper import Article
```

```python
# Spécifier les articles qui m'intéressent
urls=["https://www.20minutes.fr/high-tech/3112963-20210831-instagram-continuer-utiliser-application-faudra-desormais-renseigner-date-naissance",
     "https://www.20minutes.fr/high-tech/3112299-20210830-airbus-aimerait-remplacer-magazines-papier-bord-ecrans-numeriques",
     "https://www.20minutes.fr/high-tech/3110571-20210827-netflix-premiers-jeux-video-deja-disponibles",
     "https://www.20minutes.fr/high-tech/3110267-20210827-microsoft-faille-informatique-decouverte-cloud-milliers-entreprises-prevenues",
     "https://www.20minutes.fr/high-tech/facebook/3105931-20210820-facebook-desormais-possible-participer-reunions-realite-virtuelle",
     "https://www.20minutes.fr/high-tech/3105943-20210820-elon-musk-devoile-projet-robot-humanoide-concu-technologie-tesla"]

def url_text(url):
  article=Article(url)
  article.download()
  article.parse()
  titre=article.title
  text=article.text
  auteur=article.authors
  print(url)
  return titre,auteur,text
# List comprehension
texts = [url_text(url) for url in urls]
```
