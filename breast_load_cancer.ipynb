```
# Installation des packages
from sklearn import datasets
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```
# Télécharger la base de données
df=datasets.load_breast_cancer()
```

```
# Découvrir la base de données
print(df.DESCR)
```

```
# Préparer les données
x=df.data
y=df.target
```

```
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```
x_train,x_test,y_train,y_test=train_test_split(x_scaled , y , test_size=0.2, random_state=42)
```

```
print(x_train.shape , x_test.shape)
```

```
# Construction de réseaux de neurones
# Création de réseau de neurones
model = tf.keras.models.Sequential()
```

```
# Ajouter des couches cachées, une fonction d'activation reLU pour les couches cachées(units=15) 
# Initialisation du poids:kernel_initializer='uniform'
model.add(tf.keras.layers.Dense(units=15, activation='relu', kernel_initializer='uniform', input_dim=30))
```

```
# Ajouter la couche de sortie = 1 et la fonction d'activation sigmoïde pour la sortie(utilisé par expérience)
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))
```

```
# Ajouter une fonction coût(loss) (la différence entre le valeur prédite et la valeur réelle)
# La fonction coût(loss) met à jour les poids par une rétropropagation
# binary_crossentropy : c'est le choix entre 2 valeurs
# metrics = calcul la précision du modèle
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['binary_accuracy'])
```

```
# Synthèse du modèle construit (couche cachée et couche de sortie)
model.summary()
```

```
# Entrainer le réseau de neurone sur les données d'entrainement
# Exemple: batch_size=10 si on ne veut pas entrainer toute la base de données(parce que volumineuse)
model.fit(x_train,y_train , epochs=100)
```

```
#Tester le modèle sur les données de test
test_loss, test_accuracy = model.evaluate(x_test , y_test)
print(f'Test loss: {test_loss} et Test accuracy: {test_accuracy}')
```
