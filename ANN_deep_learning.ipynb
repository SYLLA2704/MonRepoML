**Importer les librairies**

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
```

**Importer le dataset**

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

**Preprocessing**

**Normalisation**


On va préparer un peu les données. Tout d'abord, nos données contiennent des tableaux contennant des valeurs allant de 0 à 255 ( niveau de gris ). Il est assez facile de normaliser les données en divisant chaque valeur par 255 afin de n'obtenir que des valeurs allant de 0 à 1

```python
X_train = X_train / 255.0
```

```python
X_test = X_test / 255.0
```

**Reshape**

Le seconde consiste à transformer nos images qui sont des tableaux a deux dimensions en un vecteur unique. Nous allons donc passer d"un tableau de 28 lignes, 28 colonnes à un vecteur de 784 valeurs ( 28*28).

```python
X_train = X_train.reshape(-1, 28*28)
X_test = X_test.reshape(-1, 28*28)
```

# Construction de notre ANN

Nous partons ici sur un modèle simple qui est un modèle séquentiel. https://keras.io/layers/core/

```python
model = tf.keras.models.Sequential()
```

On créé notre première couche cachée Qui est une couche dense constituée de 128 Neuronnes ( units ) avec une activation Relu (Unité de Rectification Linéaire). et en fait comme il s'agit de la première couche que nous créons nous lui indiquons le forme de nos data :-) https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense

```python
model.add(tf.keras.layers.Dense(units=128, activation='relu', input_shape=(784, )))
```

On va rajouter une couche de dropout. Ca limite l'overfitting du modèle La doc pour comprendre : https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dropout

```python
model.add(tf.keras.layers.Dropout(0.2))
```

```python
model.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
model.add(tf.keras.layers.Dropout(0.2))
```

Ici on va avoir notre couche de sortie avec 10 neurones correspondant à nos 10 classes avec une activation softmax. Un peu de théorie sur le softmax ) https://fr.wikipedia.org/wiki/Fonction_softmax

```python
model.add(tf.keras.layers.Dense(units=10, activation='softmax'))
```

```python
model.summary()
```

# Compilation et entrainement

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])
```

```python
model.fit(X_train, y_train, epochs=25)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
```

```python
# Sauvegarder le modèle
filepath = '/content/drive/MyDrive/Modèle'

tf.keras.models.save_model(
    model, filepath, overwrite=True, include_optimizer=True, save_format=None,
    signatures=None, options=None)
```
